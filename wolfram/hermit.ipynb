```
psi = {};
For[i = 0, i <= 3, i++,
    For[j = 0, j <= 3, j++,
        AppendTo[psi, u^i * v^j];
    ];
];
```

```
psi
```

```
gu = D[psi, u];
gv = D[psi, v];
guu = D[gu, u];
gvv = D[gu, v];
pts = {{0,0}, {1,0}, {1,1}, {0,1}};
```

```
h1 = Table[psi /. {u -> pts[[i,1]], v -> pts[[i,2]]}, {i,1,4}];
h2 = Table[guu /. {u -> pts[[i,1]], v -> pts[[i,2]]}, {i,1,4}];
h3 = Table[gvv /. {u -> pts[[i,1]], v -> pts[[i,2]]}, {i,1,4}];
h4 = {gv /. {v -> 0}, gu /. {u -> 1}, -gv /. {v -> 1}, -gu /. {u -> 0}};
h = Join[h1, h2, h3, h4];
```
