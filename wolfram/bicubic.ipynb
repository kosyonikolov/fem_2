```
stdBasis = Flatten[KroneckerProduct[{1,y,y^2,y^3}, {1,x,x^2,x^3}]];
stdDx = D[stdBasis, x];
stdDy = D[stdBasis, y];
stdDxDy = D[stdBasis, x, y];
pts = {{0,0}, {0,1}, {1,0}, {1,1}};
```

```
b = Table[{stdBasis, stdDx, stdDy, stdDxDy} /. {x -> pts[[i,1]], y -> pts[[i,2]]}, {i,1,4}];
h = Join[b[[1]], b[[2]], b[[3]], b[[4]]];
invH = Inverse[h];
psi = Transpose[invH] . stdBasis;
gpsi = Grad[psi, {x,y}];
```
